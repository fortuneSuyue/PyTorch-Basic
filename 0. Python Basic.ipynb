## 0. Jupyter Notebook Tips and Tricks
1. 제목을 클릭하면 파일 이름 변경 가능
2. 셀을 누른 상태에서 a를 누르면, 위에 셀 추가
3. 셀을 누른 상태에서 b를 누르면, 아래에 셀 추가
4. 셀을 누른 상태에서 d를 두번 누르면, 해당 셀 삭제
5. Shift를 누른 상태에서, 드래그 가능
6. 셀을 클릭 후 M을 누르면, Code에서 Markdown으로 전환
7. 셀을 클릭 후 Y를 누르면, Markdown에서 Code로 전환
8. 셀을 클릭 후 F를 누르면, 코드를 Find and replace
9. 셀을 클릭 후 O를 누르면, Output을 toggle
10. Ctrl + Enter를 누르면, 실행
11. Shift + Enter를 누르면, 실행 후 새로운 Code 셀 생성
12. Shift + Tap을 누르면, 객체의 doc string 출력
13. Shift + 키보드 방향키 아래를 누르면, 아래 방향으로 복수의 셀을 선택 가능
14. Shift + 카보드 방향키 위를 누르면, 위 방향으로 복수의 셀을 선택 가능
15. Shift + M을 누르면, 선택된 셀을 merge

## 8. Function

```python
def fibonacci(num):
    if num < 2:
        return num
    else:
        return fibonacci(num-1) + fibonacci(num-2)
        
print(fibonacci(3))
```

Error and Exception

```python
def tenDividedBy(num):
    try:
        return 10 / num
    except ZeroDivisionError as e:
        print("Error: cannot divide by zero", e)

tenDividedBy(0)
```

## 10. Classes and Objects

```python
class Protoss(object):
    def __init__(self, name): # initialization
        self.name = name
    def command(self):
        print("For Aiur!")
    def attack(self):
        print("Gau'ju!")
        
class Zealot(Protoss):
    def move(self):
        print("I long for combat!")
        
class Carrier(Protoss):
    def move(self):
        print("Affirmative!")
        
zealot = Zealot("zealot")
zealot.command()
zealot.attack()
zealot.move()
print("\n")

carrier = Carrier("carrier")
carrier.move()
```
